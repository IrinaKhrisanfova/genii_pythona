```python
df['Сложность'] = df['Сложность'].fillna(0)
```

```python
df['Сложность'] = df['Сложность'].replace(['Простой', 'Средний', 'Сложный'],[1, 2, 3])
```

```python
df
```

```python
df['Сложность'] = df['Сложность'].replace([0],[','])
```

```python
df
```

```python
df[['Время чтения', 'мин']] = df['Время чтения']. str.split(' ', 1 , expand= True )
```

```python
df
```

```python
df.pop('мин')
```

```python
df
```

```python
df = df.astype({'Время чтения': np.int64})
```

```python
df
```

```python
def value_to_float(x):
    if 'K' in x:
        return float(x.replace('K', '')) * 1000
    else:
        return float(x)
df['Количество просмотров'] =  df['Количество просмотров'].apply(value_to_float) 
```

```python
df
```

```python
df[['Дата публикации', 'Время публикации']] = df['Время публикации']. str.split(' ', 1 , expand= True )
```

```python
df
```

```python
df['Дата публикации'] = pd.to_datetime(df['Дата публикации'])
```

```python
df['Время публикации'] = pd.to_datetime(df['Время публикации'])
```
